```python
import sqlite3 as sql
import pandas as pd

con = sql.connect("casas.db")
con.close()
```

```python
#Con GROUP BY podemos agrupar valores unicos de una columna. 

route = "casas.db"
con = sql.connect(route)

querry = """
        SELECT habitaciones
        FROM casas
        GROUP BY habitaciones;
        """

df = pd.read_sql_query(querry, con)

df
```

```python
#Es muy útil para usarse con funciones de agregación.

querry = """
        SELECT pais, 
        COUNT(*) AS cantidad_de_ciudades
        FROM ciudades
        GROUP BY pais;
        """
df = pd.read_sql_query(querry, con)

df
```

```python
#El orden de gerarquia es FROM -> WHERE -> GROUP BY.

query = """
        SELECT id_ciudad,
        MAX(precio) AS mas_cara
        FROM casas
        GROUP BY id_ciudad;
        """
df = pd.read_sql_query(query, con)

df
```

```python
#la seleccion de grupos puede simplicarse con llamar a las columnas por su posicion en la consulta.

query = """
        SELECT id_ciudad,
        MAX(precio) AS mas_cara
        FROM casas
        GROUP BY 1;
        """

df = pd.read_sql_query(query, con)

con.close()
df
```

```python
#La cluasula HAVING se utiliza como un filtro que se utiliza en la agrupación (parecido al WHERE).

con = sql.connect("casas.db")

query = """
        SELECT id_ciudad,
        COUNT(*) AS cantidad_casas
        FROM casas
        GROUP BY id_ciudad
        HAVING cantidad_casas > 1;
        """

df = pd.read_sql_query(query, con)

df
```

```python
#La gerarquia sobre el HAVING se aplica a la que ya teniamos SELECT -> FROM -> WHERE -> GROUP BY -> HAVING -> ORDER BY

query = """
        SELECT id_ciudad,
        COUNT(*) AS cantidad_casas
        FROM casas
        WHERE precio > 100000
        GROUP BY id_ciudad
        HAVING cantidad_casas >= 1
        ORDER BY cantidad_casas DESC;
        """

df = pd.read_sql_query(query, con)

con.close()
df
```
