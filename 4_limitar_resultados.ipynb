```python
import sqlite3 as sql
import pandas as pd

con = sql.connect("casas.db")
con.close()
```

```python
#Con limit podemos limitar la cantidad de resultados que nos sale en la consulta.

con = sql.connect("casas.db")

querry = "SELECT * FROM casas LIMIT 3;"

pd.read_sql_query(querry,con) 
```

```python
#limit es muy útil conmbinado con la clausula de order by para obetener los mejores resultados de algún campo.

querry = "SELECT * FROM casas ORDER BY precio ASC LIMIT 1;"

pd.read_sql_query(querry, con)
```

```python
#Limit + order by son una buena combinación en términos de ranking.

querry = "SELECT * FROM casas ORDER BY anio_construccion ASC LIMIT 3;"

pd.read_sql_query(querry, con)
```
