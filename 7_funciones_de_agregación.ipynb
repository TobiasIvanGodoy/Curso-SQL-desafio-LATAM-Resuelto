```python
import pandas as pd
import sqlite3 as sql
```

```python
#La función max() nos devuelve el valor máximo de la columna que hayamos seleccionado.
con = sql.connect('casas.db')
querry = 'SELECT MAX(precio) FROM casas;'
                        
pd.read_sql_query(querry, con)
```

```python
#Así mismo, también esta la función MIN(), que devuelve el valor mínimo de la columna.
querry = 'SELECT MIN(precio) FROM casas;'

pd.read_sql_query(querry, con)
```

```python
#La función SUM() nos devuelve todos los valores sumados de una columna, mucho muy importante que sean números.

querry = 'SELECT SUM(precio) FROM casas;'

pd.read_sql_query(querry, con)
```

```python
#La función AVG() no devuelve el promedio.

querry = 'SELECT AVG(precio) FROM casas;'

pd.read_sql_query(querry, con)
```

```python
#Con COUNT() podemos contar la cantidad de registros de una columna ó la tabla en total.

querry = 'SELECT COUNT(*) FROM casas;'

pd.read_sql_query(querry, con)
```

```python
#Estas pueden ser convinadas con las otras claúsulas previamente vistas.

querry = 'SELECT COUNT(*) FROM casas WHERE direccion LIKE "A%"'

pd.read_sql_query(querry, con)
```
