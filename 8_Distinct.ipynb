```python
import sqlite3 as sql
import pandas as pd
```

```python
con = sql.connect('casas.db')
```

```python
#El keyword DISTINCT nos sirve para filtrar por valores únicos.

querry = 'SELECT DISTINCT habitaciones AS habitaciones_unicas FROM casas;'

pd.read_sql_query(querry, con)
```

```python
querry = 'SELECT COUNT(DISTINCT banos) AS cantidad_banos_unicos FROM casas'

pd.read_sql_query(querry, con)
```

```python
#Se pueden hacer combinaciones de unicidad si se quiere.

querry = 'SELECT DISTINCT habitaciones, banos FROM casas ORDER BY habitaciones DESC'

pd.read_sql_query(querry, con)
```

```python
con.close()
```
