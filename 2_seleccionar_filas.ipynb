```python
import sqlite3 as sql
import pandas as pd

con = sql.connect("casas.db")
con.close()
```

```python
#Con la clausula WHERE, podemos filtrar registros en la consulta.

route = "casas.db"
querry = "SELECT * FROM casas WHERE precio > 100000;"

con = sql.connect(route)

pd.read_sql_query(querry, con)
```

```python
#Todo comparador matemático es compatible.

querry = "SELECT * FROM casas WHERE precio <= 100000;"

pd.read_sql_query(querry, con)
```

```python
#Se pueden mezclar fiiltros de columnas y campos.

querry = "SELECT direccion, precio FROM casas WHERE precio > 100000;"

pd.read_sql_query(querry, con)
```

```python
#Se pueden obtener datos concretos con estos filtros.

querry = "SELECT direccion FROM casas WHERE id_casa = 2;"

pd.read_sql_query(querry, con)
```

```python
#Se puede filtrar por comparación de texto.

querry = "SELECT * FROM casas WHERE direccion = 'Calle Falsa 123';"

pd.read_sql_query(querry, con)
```

```python
#Se puede filtrar también por valores booleanos.

querry = "SELECT * FROM casas WHERE disponibilidad = TRUE;"

#Mi base de datos no cuenta con valores booleanos pero confia en que es así como se usa.
```

```python
#Se pueden fortalecer las claúsulas con operadores lógicos.

querry = "SELECT * FROM casas WHERE precio > 10 AND habitaciones < 3;"

pd.read_sql_query(querry, con)
```

```python
#Uso del OR.

querry = "SELECT * FROM casas WHERE precio > 100000 OR habitaciones = 2;"

pd.read_sql_query(querry, con)
```

```python
#Utilizando el tipo de dato 'date'.

querry = "SELECT * FROM casas WHERE fecha_de_construccion < '2022-12-31'"

#Vos confia que así se usa, es primero el año, despues el mes y por último el día.
```

```python
#Se puede usar tambien el operador BETWEEN que agarra valores entre 2 varemos.

querry = "SELECT * FROM casas WHERE precio BETWEEN 10 AND 150000"

pd.read_sql_query(querry, con)
```

```python
#Existe tambien la claúsula LIKE, que compara por caracteres coincidentes.

querry = "SELECT * FROM casas WHERE direccion LIKE 'A%'"

pd.read_sql_query(querry, con)
```

```python
#Con el comodín (%) por delante, podemos filtrar por terminación.

querry = "SELECT * FROM casas WHERE direccion LIKE '%0'"

pd.read_sql_query(querry, con)
```

```python
#También podemos filtrar si no queremos los valores nulos de alguna columna en particular.

querry = "SELECT * FROM casas WHERE direccion IS NOT NULL"

pd.read_sql_query(querry, con)
```

```python
#Así mismo, podemos también filtrar por columnas que si tengan null.

querry = "SELECT * FROM casas WHERE direccion IS NULL"

pd.read_sql_query(querry, con)
```
