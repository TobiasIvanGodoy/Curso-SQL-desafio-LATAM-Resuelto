```python
import sqlite3 as sql
import pandas as pd

#crear conecciones y cerrarlas (Mucho Muy Importante)

con = sql.connect("casas.db")
con.close()
```

```python
#Sí quiero seleccionar todas las columnas de una tabla.

route = "casas.db"
querry = "SELECT * FROM casas;"

con = sql.connect(route)
df = pd.read_sql_query(querry, con)
con.close()

df
```

```python
#Sí quiero seleccionar una columna en específico.

querry = "SELECT precio FROM casas;"

con = sql.connect(route)
pd.read_sql_query(querry, con)
```

```python
#Sí quiero seleccionar mas de una columna.

querry = "SELECT direccion, precio FROM casas;"

con = sql.connect(route)
pd.read_sql_query(querry, con)
```

```python
#Sí quiero seleccionar una columna con un alias.

querry = "SELECT direccion AS lugar  FROM casas;"

con = sql.connect(route)
pd.read_sql_query(querry, con)
```

```python
#Sí quiero seleccionar una columna con un alias descriptivo.

querry = "SELECT direccion AS 'lugar de residencia'  FROM casas;"

con = sql.connect(route)
pd.read_sql_query(querry, con)
```
