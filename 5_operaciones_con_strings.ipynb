```python
import sqlite3 as sql
import pandas as pd

con = sql.connect("casas.db")
con.close()
```

```python
#Manipular los strings de la consulta suele tener alguna útilidad.

con = sql.connect("casas.db")
querry = "SELECT UPPER(direccion) FROM casas;"

pd.read_sql_query(querry, con)
```

```python
#Así como existe upper, también esta lower.

querry = "SELECT LOWER(direccion) FROM casas;"

pd.read_sql_query(querry, con)
```

```python
#Para elimiar los espacios del inicio y final de un string poder usar la funcion TRIM.

querry = "SELECT TRIM(direccion) FROM casas;"

pd.read_sql_query(querry,con)
```

```python
#Estos comandos son iterables, por lo que se pueden llamar simultaneamente.

querry = "SELECT TRIM(LOWER(direccion)) AS direccion_limpia FROM casas;"

pd.read_sql_query(querry,con)
```

```python
#LENGTH nos sirve par aver el tamaño del string en la columna seleccionada.

querry = "SELECT direccion, LENGTH(direccion) AS dir_len FROM casas ORDER BY dir_len DESC LIMIT 3;"

pd.read_sql_query(querry,con)
```

```python
#El operador concatenar lleva un ' ' con las cuales indicas el separador de la concatenación.

querry = "SELECT precio || '-' || anio_construccion AS valor_por_antiguedad FROM casas;"

pd.read_sql_query(querry, con) 
```

```python
#El método SUBSTR() nos permite hacer una sub-sequiencia de caracteres a partir de un string. El primer parámetro es el nombre de la columna, 
#el segundo el incio, y el último es la longitud.

querry = "SELECT SUBSTR(nombre, 1, 1) AS inciales_vendedor FROM propietarios;"

pd.read_sql_query(querry, con) 
```
