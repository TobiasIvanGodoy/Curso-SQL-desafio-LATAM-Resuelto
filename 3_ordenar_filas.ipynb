```python
import sqlite3 as sql
import pandas as pd

con = sql.connect("casas.db")
con.close()
```

```python
#ORDER BY ordena por defecto en orden ascendente

con = sql.connect("casas.db")
querry = "SELECT * FROM casas ORDER BY precio;"

pd.read_sql_query(querry, con)
```

```python
#Implicitamente podemos pedir que se ordenen de manera ascendente.

querry = "SELECT * FROM casas ORDER BY precio ASC;"

pd.read_sql_query(querry, con)
```

```python
#Así mismo, se puede pedir en orden descendente.

querry = "SELECT * FROM casas ORDER BY precio DESC"

pd.read_sql_query(querry, con)
```

```python
#Los campos nulos tiene el valor mas bajo, por lo que ordenar de manera ascendente los pondrá en los primeros de la lista.
#Podemos especificar tambien si los queremos primeros o ultimos con la orden NULLS FIRST o NULLS LAST

querry = "SELECT * FROM casas ORDER BY precio NULLS LAST;"

pd.read_sql_query(querry, con)
```

```python
#Se puede también ordenar por más de una cosa a la vez.

querry = "SELECT * FROM casas ORDER BY id_ciudad ASC, precio ASC;"

pd.read_sql_query(querry, con)
```

```python
#Podemos convinarlo con otras consultas respetando las gerarquías.

querry = "SELECT * FROM casas WHERE precio >= 100000 ORDER BY id_ciudad ASC, precio ASC;"

pd.read_sql_query(querry, con)
```
